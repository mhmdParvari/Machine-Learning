```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('BostonHousing.csv')
data.head()
```

```python
data.corr()
```

```python
x = data[['indus','age']].to_numpy()
y = data['nox'].to_numpy()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
```

```python
from my_LLS import LinearLeastSquares

lls = LinearLeastSquares()
lls.fit(x_train, y_train)
m = lls.get_slope()
y_pred = lls.predict(x_test)
print(m)
print(y_pred[:20])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_test[:,1], y_test, x_test[:,0], color='red', alpha=.5)

x0, x1 = np.meshgrid(x_test[:,0], x_test[:,1])
y = x0 * m[0] + x1 * m[1]
ax.plot_surface(x1, y, x0, alpha=.5)
```
