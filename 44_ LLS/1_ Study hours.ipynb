```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,10,100)
y = x*2

for i in range(len(y)):
    if y[i]<4: y[i] += random.uniform(-y[i], 4)
    elif y[i]>16: y[i] += random.uniform(-4, 20-y[i])
    else: y[i] += random.uniform(-4,4)
    
print(y[:20])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
```

```python
from CartesianCoordinateSystem import CooSystem

sys=CooSystem(501,801, 20)
points = np.array([x_test,y_test]).T
sys.plot_points(points, (255,255,0))

for i in range(3):
    m = random.uniform(-3, 3)
    print(m)
    sys.plot_line_with_equation(-m, 1, 0, (255,0,0)) # -mx + 1y = 0
    
plt.imshow(sys.get_img())
```

```python
from my_LLS import LinearLeastSquares

lls = LinearLeastSquares()
lls.fit(x_train, y_train)
m = lls.get_slope()
m = m[0]
print(m)

sys.plot_line_with_equation(-m, 1, 0, (0,0,255)) 

plt.imshow(sys.get_img())
```

```python
from scipy import stats

reg = stats.linregress(x_train,y_train)
s = reg.slope
print(s)
sys.plot_line_with_equation(-s, 1, 0, (0,255,0))
plt.imshow(sys.get_img())
```
